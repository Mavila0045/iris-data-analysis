```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Read Iris CSV file
iris_df = pd.read_csv("Iris.csv")
```

```python
# Show the first five records
print("First five records:")
print(iris_df.head())
```

```python
# Dataframe with a sepal width greater than 4
print("Records with sepal width greater than 4:")
sepal_width_gt_4 = iris_df[iris_df['SepalWidthCm'] > 4]
print(sepal_width_gt_4)
```

```python
# Dataframe with a petal width greater than 1
print("Records with petal width greater than 1:")
petal_width_gt_1 = iris_df[iris_df['PetalWidthCm'] > 1]
print(petal_width_gt_1)
```

```python
# Records with a petal width more than 2
print("Records with petal width more than 2:")
petal_width_gt_2 = iris_df[iris_df['PetalWidthCm'] > 2]
print(petal_width_gt_2)
```

```python
# Scatter plot between sepal length and petal length
plt.figure()
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='PetalLengthCm')
plt.title('Relationship between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

The longer the sepal Length, the longer the petal length

```python
# Scatter plot with species as hue
plt.figure()
sns.scatterplot(data=iris_df, x='SepalLengthCm', y='PetalLengthCm', hue='Species')
plt.title('Relationship between Sepal Length and Petal Length with Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend(title='Species')
plt.show()
```

Generally, you can tell the species by the sepal/petal length

## Linear Regression

Model 1

```python
# a. Create y objects
y = iris_df[['SepalLengthCm']]
```

```python
# b. Create x objects
x = iris_df[['SepalWidthCm']]
```

```python
# c. Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# d. Showing first five records of all four variables/objects
print("First five records of x_train:")
print(x_train.head())

print("\nFirst five records of x_test:")
print(x_test.head())

print("\nFirst five records of y_train:")
print(y_train.head())

print("\nFirst five records of y_test:")
print(y_test.head())
```

```python
# e. Create lr object
lr = LinearRegression()
```

```python
# f. Fit both training sets into the fit method
lr.fit(x_train, y_train)
```

```python
# g. Predict x_test and store the result into y_pred object
y_pred = lr.predict(x_test)
```

```python
# h. Show first five records from actual and predicted objects
print("First five records from actual (y_test):")
print(y_test.head())

print("\nFirst five records from predicted (y_pred):")
print(pd.DataFrame(y_pred, columns=['Predicted']).head())
```

```python
# i. Find the mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

Model 2

```python
# a. Create y object
y = iris_df[['SepalLengthCm']]
```

```python
# b. Store independent variables in x
x = iris_df[['SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
```

```python
# c. Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# d. Create and fit linear regression model
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# e. Predict x_test and store the result into y_pred
y_pred = lr.predict(x_test)
```

```python
# f. Find mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

### Which model is better?

Model 2 is better with a lower mean squared error than Model 1
